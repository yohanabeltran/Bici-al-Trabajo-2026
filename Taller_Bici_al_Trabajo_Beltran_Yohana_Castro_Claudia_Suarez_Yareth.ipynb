```python
import numpy as np, pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from ucimlrepo import fetch_ucirepo

rng = np.random.default_rng(123)

# 1. Carga el dataset completo desde UCI
bike = fetch_ucirepo(id=275)  # Bike Sharing Dataset

# 2. Extrae las características (X) y el objetivo (y)
df = bike.data.original.copy()


# 3. Limpieza mínima y orden
df["dteday"] = pd.to_datetime(df["dteday"])
df = df.sort_values("dteday").reset_index(drop=True)

print(df.head())

# 4. Selección de columnas (incluye casual/registered)
df = df[['season','yr','mnth','holiday','weekday','workingday',
         'weathersit','temp','hum','windspeed','casual','registered','cnt','dteday']]

print(df.head(), df.columns.tolist())
```

## **PREGUNTA 1**
¿La distribución de cnt puede tratarse como aproximadamente normal, realiza el histrograma y el gráfico Q-Q?

```python
# 1. Visualización
import seaborn as sns
plt.figure(figsize=(14, 5))
sns.histplot(df['cnt'], kde=True)
plt.title('Histograma de cnt')
plt.show()

#2. Grafico Q-Q
plt.figure(figsize=(14, 5))
stats.probplot(df['cnt'], dist='norm', plot=plt)
plt.title('Gráfico Q-Q de cnt')
plt.show()
```

## **PREGUNTA 2**
¿Cómo co-varía cnt con temp, hum y windspeed? ¿La relación es lineal o solo monótona?. Ayuda: Calcular la correlación de Pearson y de Spearman.

```python
# Construimos nuevas columnas
columnas_relacion = ['cnt', 'temp', 'hum', 'windspeed']
df_relacion = df[columnas_relacion]
# Correlacion de person (Mide la relacion lineal entre dos variales )
person_corr = df_relacion.corr(method='pearson')
print("correlaciòn de pearson:"+str(person_corr))
#Correlacion de spearman (mide la relacion )
spearman_corr = df_relacion.corr(method='spearman')
print("correlaciòn de spearman:"+str(spearman_corr))
```

## **PREGUNTA 3**
¿Desde qué tamaño muestral el TLC justifica intervalos y pruebas basados en normalidad para la media de cnt?




El TLC dice que, para muchas distribuciones de partida (aunque sean sesgadas), cuando el tamaño de muestra  n  es suficientemente grande, la media muestral  X¯  se distribuye aproximadamente Normal.

Adicional con n ≥ 30 el TLC suele ser aceptable, pero si la variable es muy asimétrica (como cnt), se necesitan muestras más grandes.

En distribuciones sesgadas puede requerirse n > 100 o incluso > 500 para que la media se aproxime bien a la normal.

Métodos para decidir:

Empírico (bootstrap / simulación): probar distintos tamaños muestrales y verificar cuándo las medias se ven normales (skew ≈ 0, Q–Q lineal).

Teórico (Berry–Esseen): da cotas conservadoras que suelen indicar n muy grandes.

Recomendación:

Si n pequeño/mediano, usar bootstrap para intervalos de confianza.

Si n grande (≥200–500), ya es razonable usar ICs y pruebas basadas en normalidad para la media.

```python
# 1. Filtrar datos de días laborales
cnt_laborales = df[df['workingday'] == 1]['cnt']
plt.figure(figsize=(14, 5))
sns.histplot(cnt_laborales, kde=True)
plt.title('Histograma de cnt en días laborales')
plt.show()

# Valor de referencia para la prueba
valor_hipotesis = 4500
#prueba de hipotesis(media muestral de 1700=media poblacional de 4500, hipotesis nula)
#prueba alternativa (media muestral de 1700 diferente media poblacional de 4500, hipotesis alternativa)

media_muestral = cnt_laborales.mean()
desviacion_estandar = cnt_laborales.std()
n = len(cnt_laborales)
print(media_muestral, desviacion_estandar)

# 2. Realizar la prueba de hipótesis
t_statistic, p_value = stats.ttest_1samp(cnt_laborales, valor_hipotesis)

# interpretacion
```
